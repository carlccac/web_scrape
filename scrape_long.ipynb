```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

pd.set_option('display.max_rows', None)

url = 'https://en.wikipedia.org/wiki/List_of_current_members_of_the_United_States_House_of_Representatives'

df=pd.DataFrame(columns=['state','congressman','party'])

s = BeautifulSoup(requests.get(url).text, 'html.parser')

rows = s.find('table',{'class': 'wikitable sortable', 'id':'votingmembers'}).tbody.find_all('tr')
#columns = [v.text.replace('\n', '') for v in rows[0].find_all('th')]

for i in range(1, len(rows)):
    tds = rows[i].find_all('td')
    if len(tds) == 9:
        values = [tds[0].text.replace('\n', ''), tds[1].text.replace('\n', ''), tds[3].text.replace('\n', '')]
    else:
        values = [tds[0].text.replace('\n', ''), 'vacant', 'vacant']
    df = df.append(pd.Series(values, index=['state','congressman','party']), ignore_index=True)
    
print(df)

'''
sources:

Web Scrape Wikipedia Manufacture Companies Table Into a CSV File | Web Scraping with Python
https://www.youtube.com/watch?v=cdAYhLSXDj4

Web Scraping Wikipedia tables using Python
https://www.youtube.com/watch?v=ICXR9nDbudk
'''
```
